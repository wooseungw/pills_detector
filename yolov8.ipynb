```python
# Load YOLOv8n, train it on COCO128 for 3 epochs and predict an image with it
from ultralytics import YOLO

model = YOLO('yolov8n.pt')  # load a pretrained YOLOv8n detection model
model.train(data='C:\Workspace\pills_detector\dataeset\data.yaml',verbose=True, epochs=150)  # train the model
```

```python
model.val()
```

```python
model = YOLO('C:\Workspace/pills_detector/runs/detect/train7/weights/best.pt')
results = model('C:\Workspace\pills_detector\medicine_origin\Amocla Tab..jpg')
```
